```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Get unique player into a list
total_players = purchase_data["SN"].unique()

#Count the number
num_total_players = len(total_players)

#Put into a dataframe
num_total_players_df = [{"Total Players": num_total_players}]
num_total_players_df = pd.DataFrame(num_total_players_df)
num_total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Number of unique item
unique_item = purchase_data["Item Name"].unique()
num_unique_item = len(unique_item)

#Average Prices
average_prices = purchase_data["Price"].mean()
average_prices = round(average_prices, 2)

#Number of Purchases
num_purchase = purchase_data["Purchase ID"].count()

#Total Revenue
total_revenue = purchase_data["Price"].sum()

#Summarize into a dataframe
data_summary_1_df = [{"Number of Unique Items": num_unique_item
                   , "Average Price": average_prices
                   , "Number of Purchases": num_purchase
                   , "Total Revenue": total_revenue }]
data_summary_1_df = pd.DataFrame(data_summary_1_df)
data_summary_1_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Get the unique data
purchase_data_gender = purchase_data[["SN","Gender"]]
purchase_data_gender = purchase_data_gender.drop_duplicates(subset=["SN","Gender"], keep= "last")

#Get numbers from dataframe
male_players = purchase_data_gender.loc[(purchase_data_gender["Gender"] == "Male")] 
num_male_players = len(male_players)

female_players = purchase_data_gender.loc[(purchase_data_gender["Gender"] == "Female")]
num_female_players = len(female_players)

num_other_players = num_total_players - num_male_players - num_female_players

#Get precentage
per_male_players = round((num_male_players/num_total_players)*100, 2)
per_female_players = round((num_female_players/num_total_players)*100, 2)
per_other_players = round((num_other_players/num_total_players)*100, 2)

#Summarize into a dataframe
data_summary_2_df = pd.DataFrame([(num_male_players, per_male_players)
                                 ,(num_female_players, per_female_players)
                                 ,(num_other_players, per_other_players)],
                    index = ["Male", "Female", "Other / Non-Disclosed"],
                    columns = ["Total Count","Percentage of Players"])
data_summary_2_df 
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Get a copy of original data
age_data = purchase_data

#Create bins
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 1000]
group_name = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Group
age_data["Age Range"] = pd.cut(age_data["Age"], bins, labels=group_name)

#Count
age_data = age_data.groupby(["Age Range"])
age_data_sum = age_data["SN"].nunique()

#Get into a frame
age_data_df = pd.DataFrame(age_data_sum)
age_data_df["Percentage of Players"] = round((age_data_df["SN"] / num_total_players)*100,2)
age_data_df = age_data_df.rename(columns={"SN":"Total Count"})

age_data_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Get a copy of age data
age_details = age_data

#Count
total_purchase_age = age_details["Price"].sum()
total_item_age = age_details["Item ID"].count()

#Put into frame
age_details_df = pd.DataFrame(total_item_age)
age_details_df = age_details_df.rename(columns={"Item ID":"Purchase Count"})
age_details_df["Total Purchase Value"] = total_purchase_age
age_details_df["Average Purchase Price"] = round(total_purchase_age / total_item_age,2)
age_details_df["Avg Total Purchase per Person"] = round(total_purchase_age / age_data_sum,2)

#$
age_details_df["Total Purchase Value"] = age_details_df["Total Purchase Value"].map("${:.2f}".format)
age_details_df["Average Purchase Price"] = age_details_df["Average Purchase Price"].map("${:.2f}".format)
age_details_df["Avg Total Purchase per Person"] = age_details_df["Avg Total Purchase per Person"].map("${:.2f}".format)

age_details_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Get data
purchase_datails = purchase_data.groupby(["SN"])
total_spent_player = purchase_datails["Price"].sum()
items_per_player = purchase_datails["Price"].count()

#Put into dataframe
purchase_detail_df = pd.DataFrame(total_spent_player)
purchase_detail_df["Purchase Count"] = items_per_player
purchase_detail_df["Average Purchase Price"] = round(total_spent_player / items_per_player,2)
purchase_detail_df = purchase_detail_df.rename(columns = {"Price":"Total Purchases"})

#Sorting
purchase_detail_df.sort_values(by='Total Purchases', ascending=False, inplace=True)
purchase_detail_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Get data
popular_item = purchase_data.groupby(["Item ID", "Item Name"])
total_purchases = popular_item["Price"].sum()
num_items_purchased = popular_item["Item ID"].count()

#Calculate and put into dataframe
popular_item_df = pd.DataFrame(total_purchases)
popular_item_df["Purchase Count"] = num_items_purchased
popular_item_df = popular_item_df.rename(columns = {"Price":"Total Purchase Value"})
popular_item_df["Item Price"] = popular_item_df["Total Purchase Value"] / num_items_purchased

#
popular_item_df = popular_item_df.sort_values(by="Purchase Count", ascending=False)
popular_item_df = popular_item_df[["Purchase Count","Item Price","Total Purchase Value"]]

popular_item_df.head(5)
```
